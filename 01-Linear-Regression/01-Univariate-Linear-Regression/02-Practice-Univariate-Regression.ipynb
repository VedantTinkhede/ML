```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv(r"C:\Users\ACER\Desktop\ML\01-Linear-Regression\01-Univariate-Linear-Regression\05-canada_per_capita_income.csv")
data
```

```python
plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter(data['year'], data['per capita income (US$)'], color = 'red')
```

```python
reg = linear_model.LinearRegression()
reg.fit(data[['year']], data['per capita income (US$)'])
```

```python
reg.predict([[2020]])
```

```python
plt.xlabel('Year')
plt.ylabel('Per capita income (in US$)')
plt.scatter(data['year'], data['per capita income (US$)'], color = 'red')
plt.plot(data['year'], reg.predict(data[['year']]))
```
