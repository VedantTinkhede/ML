# Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

# Importing the data

```python
df = pd.read_csv(r"C:\Users\ACER\Desktop\ML\01-Simple-Linear-Regression\homeprices.csv")
df
```

# Visualising the data on a scatter plot

```python
%matplotlib inline
plt.xlabel('area(sq. ft.)')
plt.ylabel('price(sq. ft.)')
plt.scatter(df.area, df.price, color = "red", marker = '+')
```

# Training the linear model

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

# Predicting the price of an arbitary area

```python
reg.predict([[5000]])
```

## Considering the line: y = m * x + c, m is the coefficient and c is the intercept

```python
reg.coef_  
```

```python
reg.intercept_
```

# Visualising the data with the fitted line

```python
%matplotlib inline
plt.xlabel('area(sq. ft.)')
plt.ylabel('price(sq. ft.)')
plt.scatter(df.area, df.price, color = "red", marker = '+')
plt.plot(df.area, reg.predict(df[['area']]))
```
